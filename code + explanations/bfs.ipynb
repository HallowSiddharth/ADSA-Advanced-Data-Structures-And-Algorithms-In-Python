# **Breadth First Search (BFS)**

### Let's take the following graph as an example


```python
# Constructing this graph first
graph={
    "A":["B","C"],
    "B":["D","E","F"],
    "C":["G"],
    "D":[],
    "E":[],
    "F":["H"],
    "G":["I"],
    "H":[],
    "I":[]
}
```

### **BFS Algorithm**

> 1. A queue data structure is used (First in First Out).
> 2. Once a node is visited, it's adjacent nodes are added to the end of the queue.
> 3. So basically, pop the first element from the queue, mark it as visited and add it's adjacent ones to the end.
> 4. Repeat till the queue is empty.

```python
#BFS Implementation

def bfs(graph,node):
    queue=[]
    visited=[]
    queue.append(node)
    while queue!=[]:
        element=queue.pop(0)
        visited.append(element)
        queue.extend(graph[element])
    print(" ".join(visited))


#driver code to test it out
bfs(graph,"A") #from level 1
bfs(graph,"B") #from level 2
bfs(graph,"D") #from level 3
```

> **Note: in the last case, we only get D as output and not D E F G H I because D is not adjacent to any node (i.e D is a leaf node).**

> In other words, same level works only starting from the given node's adjacent nodes. If there are no adjacent nodes, there is nothing to search for, hence the result.
